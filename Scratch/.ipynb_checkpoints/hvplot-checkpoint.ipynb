```python
import pandas as pd
import hvplot.pandas
```

```python
url = "https://raw.githubusercontent.com/resbaz/r-novice-gapminder-files/master/data/gapminder-FiveYearData.csv"
df = pd.read_csv(url)
df.head()
```

## Categorical

```python
# bar plot
df.hvplot.bar(x='continent', y='lifeExp')
```

```python
df.hvplot.barh(x='continent', y='lifeExp')
```

```python
df.hvplot.box(y='lifeExp')
```

```python
df.hvplot.box(y='lifeExp', by='continent')
```

```python
## for interactivity, visualize by continents with one box per country. Widget for continent
df.hvplot.box(y='lifeExp', by='country', width=1500, rot=90)
```

```python
## for interactivity, visualize by continents with one box per country. Widget for continent
df.hvplot.box(y='gdpPercap', by='continent')
```

```python
df.hvplot.scatter(x='year', y='lifeExp')
```

```python
df.hvplot.scatter(x='year', y='lifeExp', color='continent')
```

```python
df.hvplot.box(y='lifeExp', by='year')
```

```python
# does not work
df.hvplot.box(y='lifeExp', by='year', color='continent')
```

```python
df.hvplot.box(y='lifeExp', by=['year', 'continent'], color='continent', width=1500, rot=90)
```

```python
df.hvplot.box(y='lifeExp', by=['year', 'continent']).groupby('continent')
```

```python
df.hvplot.scatter(x='year', y='lifeExp', by='continent', subplots=True)
```

```python
# linked axes. to unlink, use shared_axes=False
df.hvplot.scatter(x='year', y='lifeExp', by='continent', subplots=True).cols(1)
```

```python
df.hvplot.hist(y='lifeExp')
```

```python
df.hvplot.hist(y='lifeExp', by='continent')
```

```python
df.hvplot.hist(y='lifeExp', by='continent', subplots=True).cols(1)
```
