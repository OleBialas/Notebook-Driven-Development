# Interactive Plots in Jupyter Notebook with hvPlot

Being able to interact with your figures and data tables directly makes it easy and fun to do Exploratory Data Analysis (EDA). `hvPlot` can be used to make visualizations 

```python
url = "https://raw.githubusercontent.com/resbaz/r-novice-gapminder-files/master/data/gapminder-FiveYearData.csv"
df = pd.read_csv(url)
df
```

## Section 1: Widgets

```python
df[['age', 'sex']].hvplot.hist(groupby='sex', width=500)
```

```python
df[['age', 'pclass']].hvplot.hist(groupby='pclass', width=500)
```

```python
df[['age', 'pclass', 'sex']].hvplot.hist(groupby=['pclass', 'sex'], width=500)
```

```python
df[['age', 'pclass', 'sex']].hvplot.hist(groupby=['pclass', 'sex'], width=200) + df[['fare', 'pclass', 'sex']].hvplot.hist(groupby=['pclass', 'sex'], width=200)
```
